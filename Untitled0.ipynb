```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np


x=np.array([[1],[2],[3],[4],[5]])
y=np.array([0,0,0,1,1])

clf=KNeighborsClassifier(n_neighbors=4)
clf.fit(x,y)

print(clf.predict([[9]]))
```
